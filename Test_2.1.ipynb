```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import ScalarFormatter, MaxNLocator
import seaborn as sns
from ipywidgets import interact, widgets

df_gpp_lavoratori = pd.read_excel(r"C:\Users\uervi\Desktop\analisi_microdati\df_gpp_lavoratori_xlsx.xlsx")

df_gpp_lavoratori['anno'] = df_gpp_lavoratori['anno'].astype('category')
df_gpp_lavoratori["genere"] = df_gpp_lavoratori["genere"].astype('category')
df_gpp_lavoratori['regione'] = df_gpp_lavoratori['regione'].astype('category')
df_gpp_lavoratori['codsettore'] = df_gpp_lavoratori['codsettore'].astype('object')
df_gpp_lavoratori['descrizione_settore'] = df_gpp_lavoratori['descrizione_settore'].astype('category')
df_gpp_lavoratori['ggp'] = df_gpp_lavoratori['ggp'].astype(str)
df_gpp_lavoratori['numero_lavoratori'] = df_gpp_lavoratori['numero_lavoratori'].astype(int)

df_generale_ggp = df_gpp_lavoratori.groupby(
    ['anno', 'ggp', 'genere'],
    as_index=False
    )['numero_lavoratori'].sum()

#print(df_generale_ggp.head())

def mostra_tabella_gpp(gruppo_prof):
    dati_filtrati = df_generale_ggp[df_generale_ggp['ggp'] == gruppo_prof]
    
    d_tabella_gpp = dati_filtrati.pivot(
        index = 'anno',
        columns = 'genere',
        values = 'numero_lavoratori'
    )
    plt.figure(figsize=(18, 12),)
    
    tabella_gpp = plt.table(
        cellText=d_tabella_gpp.values,
        rowLabels=d_tabella_gpp.index,
        colLabels=d_tabella_gpp.columns,
        cellLoc='center',
        loc='center',
        bbox=[0, 0, 1, 1],
    )
    
    for key, cells in tabella_gpp.get_celld().items():
        cells.set_fontsize(16)
        
    plt.axis('off')
    
    plt.title(f'Tabella divisa per genere del gruppo: {gruppo_prof}', fontsize=14)
    
    plt.show()

gpp_dropdown = widgets.Dropdown(
    options=df_generale_ggp['ggp'].unique(),
    description='Gruppo professionale',
    values=df_generale_ggp['ggp'].unique()[0],
    disabled=False,
)

interact(mostra_tabella_gpp, gruppo_prof=gpp_dropdown)

def mostra_grafico_gpp(gruppo_prof):
    dati_filtrati = df_generale_ggp[df_generale_ggp['ggp'] == gruppo_prof]
    datif_maschile = dati_filtrati[
        dati_filtrati['genere'] == 'M']
    datif_femminile = dati_filtrati[
        dati_filtrati['genere'] == 'F']

    anni = dati_filtrati['anno'].unique()
    
    lavoratori_maschile = [
        datif_maschile[datif_maschile['anno'] == anno]
        ['numero_lavoratori'].sum() 
        for anno in anni]
    lavoratori_femminile = [
        datif_femminile[datif_femminile['anno'] == anno]
        ['numero_lavoratori'].sum() 
        for anno in anni]
    
    x = np.arange(len(anni))
    offset = 0.2
    
    plt.figure(figsize=(14, 8))
    
    plt.plot(
        x + offset,
        lavoratori_maschile,
        label='M',
        color='blue',
        marker='o',
        alpha=0.8)
    
    plt.plot(
        x - offset,
        lavoratori_femminile,
        label='F',
        color='red',
        marker='o',
        alpha=0.8)
    
    plt.bar(
        x + offset,
        lavoratori_maschile,
        color='blue',
        alpha=0.4,
        width=0.3,)
    
    plt.bar(
        x - offset,
        lavoratori_femminile,
        color='red',
        alpha=0.4,
        width=0.3,)
    
    plt.xticks(x, anni)
    plt.title(f'Numero di lavoratori di tipo: {gruppo_prof}', fontsize=14)
    plt.xlabel('Anno', fontsize=12)
    plt.ylabel('Numero di lavoratori', fontsize=12)
    plt.grid(True)
    plt.legend(title='Genere', fontsize=10)
    plt.gca().yaxis.set_major_formatter(ScalarFormatter())
    plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
        
    plt.show()

    
interact(mostra_grafico_gpp, gruppo_prof=gpp_dropdown)
```
